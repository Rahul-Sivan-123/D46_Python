```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('myexcel.csv')
df
```

```python
df['Height']=np.random.randint(150,180,size=458)
df
```

```python
# 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
```

```python
distribution=df['Team'].value_counts()
print(distribution)

# Line Chart Visualization
distribution.plot(kind='line',title='Employee Distribution Across Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
percentage_split=(distribution/len(df))*100
print(percentage_split)

teams=distribution.index.tolist()
counts=distribution.values.tolist()

# Pie Chart Visualization
plt.figure(figsize=(8,6))
plt.pie(counts,labels=teams,autopct='%1.1f%%',colors=plt.cm.Paired.colors,startangle=90)
plt.title('Employee Percentage Distribution Across Team')
plt.show()
```

```python
# 2. Segregate employees based on their positions within the company. 
```

```python
position=df['Position'].value_counts()
position_summary=pd.DataFrame({
    'Position':position.index,'Employee count':position.values})
print(position_summary)

# Line chart Visualization
xaxis=position.index.tolist()
yaxis=position.values.tolist()
plt.plot(xaxis,yaxis,marker='s',color='r')
plt.xlabel('Position')
plt.ylabel('Employee Count')
plt.title('Position vs Employee_Count')
plt.grid()
plt.show()
```

```python
# 3. Identify the predominant age group among employees. (2 marks) 
```

```python
df['Age'].unique()
```

```python
age_bins=[20,25,30,35,40]
age_labels=['20-25','26-30','31-35','36-40']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
age_group_distribution=df['Age Group'].value_counts().sort_index()
print(age_group_distribution)

# Bar chart Visualization
age_group_df = age_group_distribution.reset_index()
age_group_df.columns = ['Age Group', 'Count']

plt.figure(figsize=(8, 6))
plt.bar(age_group_df['Age Group'], age_group_df['Count'], color='teal', alpha=0.7)
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employees by Age Group')
plt.show()
```

```python
 # 4. Discover which team and position have the highest salary expenditure. 
```

```python

team_salary=df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
team_salary1=df.groupby('Team')['Salary'].sum().reset_index()
team_salary1.colums=['Team','Salary']
highest_salary = team_salary1.sort_values('Salary',ascending=False).iloc[0]
print('Total Salary by Team:\n',team_salary1)
print('Team having Highest Salary Expenditure is:\n',highest_salary)

# Bar Chart Visualization
plt.figure(figsize=(10, 6))
team_salary.plot(kind='bar',title=' Total Salary By Team')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
position_salary=df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
position_salary1=df.groupby('Position')['Salary'].sum().reset_index()
position_salary1.columns = ['Position', 'Salary']
high_salary =position_salary1.sort_values(by='Salary',ascending=False).iloc[0]
print('Total Salary by Position:\n',position_salary1)
print('Position having Highest Salary Expenditure is:\n',high_salary)

# Bar Chart Visualization
plt.figure(figsize=(10, 6))
position_salary.plot(kind='bar',title=' Total Salary By Position',color='g')
plt.ylabel('Salary')
plt.xlabel('Team Position')
plt.legend()
plt.grid()
plt.show()
```

```python
# 5. Investigate if there's any correlation between age and salary, and represent it visually. 
```

```python
df.corr(numeric_only=True)

# Heat Map Visualization
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='mako',linewidth=0.8)
plt.title( 'Heat Map')
plt.show()
```

```python
age='Age'
salary='Salary'

correlation_value = df[age].corr(df[salary])
print(f"Correlation between {age} and {salary}: {correlation_value}")

# Scatter plot to visualize correlation
plt.figure(figsize=(8, 6))
plt.scatter(df[age], df[salary], alpha=0.8,color='b')
plt.title(f"Correlation between {age} and {salary}: {correlation_value:.2f}")
plt.xlabel(age)
plt.ylabel(salary)
plt.grid(True)
plt.show()
```
